```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent = [
        'the glass of milk',
        'the glass of juice',
        'the cup of tea',
        'I am a good boy',
        'I am a good developer',
        'understand the meaning of words',
        'your videos are good'
]
```

```python
# defining the vocab size
vocab = 10000
```

```python
# one hot representation

one_hot_repr = [one_hot(words, vocab) for words in sent]
one_hot_repr
# here we can see what kind of index that word was given out like in one hot encoding
```

```python
# word embedding representation

from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras import Input
import numpy as np
```

```python
sent_lenght = 8

# in order to make all of the sentences same lenght we use pad_sequence
embedded_docs = pad_sequences(one_hot_repr, padding='pre', maxlen=sent_lenght)
print(embedded_docs)
```

```python
# feature representation

dim = 10
```

```python
# model init

model = Sequential()
model.add(Input(sent_lenght))
model.add(Embedding(vocab, dim))
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
