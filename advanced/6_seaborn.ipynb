```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

```python
# sns is a library similar to matplotlib just more adapted for data science and forecasting

sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("Line Plot of Total bill by size")
plt.show()
```

```python
# barplots are also available, but its different since its provide more statistical information (in this case we see CIs)

sns.barplot(x='day',y='total_bill',data=tips)
plt.title('Bar Plot of Total Bill By Day')
plt.show()
```

```python
# box plots are also available, but here we are provided with median aswell 
sns.boxplot(x="day",y='total_bill',data=tips)
```

```python
# violin plots are also possible
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
# histogram
sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
# KDE Plot
sns.kdeplot(tips['total_bill'],fill=True)
```

```python
# pairplot
sns.pairplot(tips)
```

```python
# also useful correlation plots are included in sns library
corr = tips[['total_bill','tip','size']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

```python
# new dataset
import pandas as pd
import os 

full_path = os.path.join('datasets','sales_data.csv')

sales_df = pd.read_csv(full_path)
sales_df.head()
```

```python
# Plot total sales by product
plt.figure(figsize=(10,6))
sns.barplot(x='Product Category',y="Total Revenue",data=sales_df,estimator=sum)
plt.title('Total Sales by Product')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.show()
```

```python
# plots for regions can be provided aswell
plt.figure(figsize=(10,6))
sns.barplot(x='Region',y="Total Revenue",data=sales_df,estimator=sum)
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```
